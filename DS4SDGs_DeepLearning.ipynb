## 1.2.1 Typical Neural Network Architecture:
The typical neural network architecture consists of several layers.

### 1.2.1 Input Layer
The input layer is the initial layer through which the neural network receives its input X, which represents the data from which it learns.

The number of neurons (nodes) in this layer is typically equal to the number of features in the input data, X.

Ideally, each input neuron represents one element in the vector x.

Basically, the input layer prepares the data for further processing by the subsequent layers.

### 1.2.2 Output Layer
This is the last layer of the neural network which outputs a vector y representing the result that the neural network came up with. The entries in this vector represent the values of the neurons in the output layer. Each neuron in the last layer would represent a different class.

For instance, for a binary classification problem, you might have one output neuron, while for a multi-class classification problem, you might have as many neurons as there are classes.

### 1.2.3 Hidden Layer
In order to obtain a prediction vector y, the network must perform certain mathematical operations. These operations are performed in the layers between the input and output layers which are called Hidden Layers.

These layers, aptly named because they are not directly observed from the input or output data, are pivotal in enabling deep learning models to represent and learn complex functions.

üí° **Fun Fact**  
The "deep" in deep learning or deep neural network usually refers to the multiple hidden layers found in the neural networks.

There are several types of hidden layers including: Dense (fully connected), convolutional layers, recurrent layers, normalization layers, etc.

1. **Dense (Fully Connected) Layers:**  
   These are the most common type where every neuron in the layer is connected to every neuron in the subsequent layer. The dense layer is a linear operation, but when combined with a non-linear activation function, it provides the network the capability to learn from the error and make adjustments, which is essential for learning complex patterns.

2. **Convolutional Layers:**  
   Primarily used in convolutional neural networks (CNNs), these layers are specifically designed to recognize spatial hierarchies in data, making them excellent for image recognition tasks. They work by sliding small windows (or kernels) over the input data to produce feature maps, capturing local patterns.

3. **Recurrent Layers:**  
   Found in recurrent neural networks (RNNs), these layers have loops that allow information to be passed from one step of the network to the next. This architecture is particularly beneficial for tasks where temporal dynamics and sequence recognition are crucial, like in natural language processing or time series prediction.

4. **Normalization Layers:**  
   These layers, such as Batch Normalization, adjust and scale the activations. By normalizing the layer's inputs, they can help maintain a stable mean and variance, improving the overall training speed and performance of deep networks by mitigating the problem of internal covariate shift.

## 1.2.3 Learning Process of a Neural Network:
A neural network learns by adjusting its weights based on the error in its predictions. The learning process consists of two main steps: **Forward Propagation** and **Backpropagation**.

### Forward Propagation:
1. **Input Layer:** The input layer receives the input data and passes it to the next layers.
2. **Hidden Layers:** The data is passed through one or more hidden layers, where it is processed using weights, biases, and activation functions.
3. **Output Layer:** The final layer outputs the network's prediction or result.

At this stage, the neural network makes a prediction based on the given input.

### Backpropagation (Geri Yayƒ±lƒ±m):
Backpropagation is the process by which the neural network learns from the error in its prediction.

1. **Error Calculation:** After forward propagation, the network calculates the error, which is the difference between the predicted value and the actual value (the ground truth).
2. **Gradient Calculation:** The network calculates the gradients (partial derivatives) of the error with respect to each weight in the network.
3. **Update Weights:** Using the gradients, the weights are adjusted to minimize the error. This is usually done using an optimization algorithm like **Gradient Descent**.

Backpropagation helps the network adjust the weights in the right direction to improve its predictions. This process is repeated for many iterations (epochs), allowing the network to gradually learn from the data and make more accurate predictions over time.

### Simple Summary:
- **Forward Propagation** is where the network makes predictions.
- **Backpropagation** is how the network learns by adjusting its weights based on the error in its predictions.

## 1.2.4 Loss Function:
Loss functions measure the difference between a neural network's predictions and the actual values. The goal of training is to minimize this loss for better accuracy.
![](https://miro.medium.com/v2/resize:fit:1400/0*QLXFJpiWv2qt2qcs.gif)

### 1. **Mean Squared Error (MSE) Loss**
- **Definition:**  
  MSE calculates the average of the squared differences between the actual values ($y_i$) and predicted values ($\hat{y}_i$).

- **Formula:**  
  $$
  \text{MSE} = \frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2
  $$

- **Usage:**  
  Used for regression tasks such as predicting house prices or temperatures.

- **Key Point:**  
  Squaring the errors amplifies large deviations, heavily penalizing significant errors.

![](https://miro.medium.com/v2/resize:fit:1400/1*n37x4cBieB5LBmOZCioSBw.gif)

---

### 2. **Cross Entropy Loss**
- **Definition:**  
  Measures the dissimilarity between predicted probabilities ($\hat{y}_i$) and true labels ($y_i$).

- **Formula (Binary Classification):**  
  $$
  \text{Cross Entropy Loss} = -\frac{1}{n} \sum_{i=1}^n \Big(y_i \cdot \log(\hat{y}_i) + (1 - y_i) \cdot \log(1 - \hat{y}_i)\Big)
  $$

- **Usage:**  
  Standard for classification problems like image recognition or text analysis.

- **Key Point:**  
  Encourages the model to produce confident, correct predictions and penalizes incorrect overconfidence.

![](https://pmirla.github.io/assets/entropy/entropy_ex2.gif)

---

### Summary:
- **MSE:** Best for regression, minimizing the numerical error.  
- **Cross Entropy:** Best for classification, focusing on probability distributions.


## 1.2.5 Gradient Descent

### What is Gradient Descent?
Gradient Descent is an optimization algorithm used to minimize the loss function of a machine learning model. It iteratively updates the model parameters by moving in the direction of the steepest descent (negative gradient) to find the minimum value of the loss function.


---

### Gradient Descent Formula
The parameter $\theta$ is updated using the following rule:
$$
\theta = \theta - \alpha \frac{\partial L}{\partial \theta}
$$

Where:
- $\theta$: Model parameter (e.g., weights in neural networks)  
- $\alpha$: Learning rate (controls step size)  
- $\frac{\partial L}{\partial \theta}$: Gradient of the loss function $L$ with respect to $\theta$

---

### The Process
1. **Initialize Parameters:** Start with random values for $\theta$.  
2. **Compute Gradient:** Calculate $\frac{\partial L}{\partial \theta}$, the slope of the loss function with respect to $\theta$.  
3. **Update Parameters:** Adjust $\theta$ using the update rule above.  
4. **Repeat:** Continue updating until the loss converges to a minimum or meets a stopping criterion.

---

### Learning Rate ($\alpha$)
The learning rate determines the size of the steps taken towards the minimum:
- **Small $\alpha$:** Slow convergence but more stable.  
- **Large $\alpha$:** Faster progress but may overshoot or diverge.  

Choosing an appropriate learning rate is crucial for the success of Gradient Descent.

---

### Variants of Gradient Descent
1. **Batch Gradient Descent:**  
   Computes the gradient using the entire dataset. It is stable but computationally expensive for large datasets.

2. **Stochastic Gradient Descent (SGD):**  
   Uses a single data point to update the parameters. It is faster but less stable, introducing noise into the optimization process.

3. **Mini-batch Gradient Descent:**  
   Uses small batches of data points, balancing stability and speed.

## 1.2.6 Curse of Dimensionality

Before building a neural network, it is essential to understand a critical challenge in high-dimensional data analysis: the **Curse of Dimensionality**.

This term, coined by Richard Bellman in the 1960s, refers to the exponential increase in computational complexity and data sparsity as the number of dimensions (features) grows.

---

### Key Impacts of the Curse of Dimensionality

1. **Data Sparsity**  
   In high-dimensional spaces, data becomes sparse. To adequately sample all feature combinations, the required data grows exponentially.  
   - **Example**:  
     If 10 samples suffice for a single feature, you'd need:
       - 100 samples for 2 features  
       - 1,000 samples for 3 features  
       - And so on.  
   Without enough data, models struggle to generalize effectively.

2. **Distance Measures Lose Meaning**  
   Many machine learning algorithms rely on distance measures (e.g., Euclidean distance). However, in high dimensions:
   - Distances between points become nearly uniform.  
   - Clustering and nearest neighbor searches become less reliable.

3. **Overfitting**  
   Higher dimensions increase the model's flexibility, making it easier to fit noise rather than patterns in the data.  
   - **Result**: The model performs well on the training set but poorly on unseen data.

4. **Computational Complexity**  
   Algorithms that work efficiently in low dimensions can become computationally intractable in high-dimensional spaces.  
   - **Example**: Matrix operations (e.g., inversions) have a complexity that grows with dimensionality.

5. **Hughes Phenomenon**  
   Hughes (1968) observed that with a fixed number of training samples:
   - The predictive power of a classifier initially improves as dimensions increase.  
   - After a certain threshold, performance deteriorates due to data sparsity and overfitting.

---

### Addressing the Curse of Dimensionality
1. **Feature Selection**  
   Identify and retain only the most relevant features for your task.  

2. **Dimensionality Reduction**  
   Techniques like **Principal Component Analysis (PCA)** or **t-SNE** can project high-dimensional data into lower-dimensional spaces while preserving critical patterns.

3. **Regularization**  
   Apply regularization techniques (e.g., L1, L2) to penalize model complexity and prevent overfitting.

4. **Increase Data Availability**  
   Collect more data to combat sparsity and improve model generalization.

---

### Conclusion
The Curse of Dimensionality poses significant challenges in high-dimensional data processing, from increased sparsity to computational bottlenecks. Understanding its implications allows us to take proactive steps like feature selection, dimensionality reduction, and careful model regularization to mitigate its effects.

## Model Structure and Layers Used

This deep learning model creates a convolutional neural network (CNN) on **TensorFlow** using the **Keras** library. The model is designed specifically for image classification problems. The layers and functions used are explained below.

### Layers and Parameters

1. **Input Layer**  
   - Defines the shape of the input data the model will receive.  
   - **shape=(224, 224, 3)**: Indicates that the input images are 224x224 pixels with 3 color channels (RGB).

2. **Conv2D (Convolutional Layer)**  
   - **Number of Filters (32)**: 32 filters are used in this layer.  
   - **Filter Size (3x3)**: Operations are performed on a 3x3 area of the image.  
   - **Activation Function**:  
     - `relu`: Helps learn nonlinear relationships by setting negative values to zero.  
     - Alternatives: `sigmoid` (for binary classification) or `tanh`.

3. **MaxPooling2D (Max Pooling Layer)**  
   - **Pooling Size (2x2)**: The maximum value from each section is selected, reducing the feature map.  
   - Reduces the computational load of the model and improves performance.

4. **Flatten (Flattening Layer)**  
   - Converts 2D image data into a 1D vector.  
   - This step is necessary for working with fully connected layers.

5. **Dense (Fully Connected Layer)**  
   - **Number of Neurons (128)**: Enables the learning of more complex features.  
   - **Activation Function**: `relu`.

6. **Dropout (Dropout Layer)**  
   - **Rate (0.5)**: Disables random neurons during each training step to prevent overfitting.

7. **Output Layer**  
   - **Number of Neurons (2)**: Indicates that the model will classify two classes.  
   - **Activation Function**:  
     - `softmax`: Produces probabilities for each class and is used for multi-class classification.  
     - Alternative: `sigmoid` (for binary classification).

### Model Compilation
- **Optimization Algorithm**:  
  - `adam`: A fast optimization algorithm that automatically adjusts the learning rate.  
  - Alternative: `sgd` (Stochastic Gradient Descent).  
- **Loss Function**:  
  - `sparse_categorical_crossentropy`: For multi-class classification problems.  
  - Alternative: `categorical_crossentropy` (when labels are one-hot encoded).  
- **Metric**: `accuracy`.

### Model Summary
The model structure, connections between layers, and information like the number of parameters can be summarized using `model.summary()`.


````python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Input

model = Sequential([
    Input(shape=(224, 224, 3)),
    Conv2D(32, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(2, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()
````

# Creating a MobileNetV2-Based Deep Learning Model

This code creates an image classification model using the **MobileNetV2** architecture. The base model (pre-trained model) is used with transfer learning, and custom fully connected layers are added.

### 1. **MobileNetV2 Base Model**
```python
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False
```
*   **`weights='imagenet'`**: MobileNetV2 is initialized with pre-trained weights from the ImageNet dataset.
*   **`include_top=False`**: The final classification layer is excluded, and the model is used only for feature extraction.
*   **`input_shape=(224, 224, 3)`**: The input images must be 224x224 pixels in size and have 3 color channels (RGB).
*   **`base_model.trainable = False`**: The weights of the base model are frozen, meaning these layers will not be retrained.
*   **`GlobalAveragePooling2D()`**: Reduces the feature maps to a single dimension, speeding up learning with fewer parameters.
*   **`Dense(128, activation='relu')`**: A fully connected layer with 128 neurons is added, using **ReLU activation**.
*   **`Dropout(0.5)`**: A dropout rate of 50% is applied to prevent overfitting by randomly deactivating neurons during training.
*   **`Dense(1, activation='sigmoid')`**: The output layer uses a sigmoid activation function for binary classification.
*   **`optimizer='adam'`**: The Adam optimization algorithm is used.
*   **`loss='binary_crossentropy'`**: The binary cross-entropy loss function is suitable for binary classification problems.
*   **`metrics=['accuracy']`**: The model performance is evaluated using accuracy as the metric.



````python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout

base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.5)(x)
predictions = Dense(1, activation='sigmoid')(x)
model = Model(inputs=base_model.input, outputs=predictions)

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
````

- **`monitor='val_loss'`**: Early stopping monitors the validation loss. If the validation loss does not improve for the specified duration, training is stopped.
- **`patience=3`**: If the validation loss does not improve for 3 consecutive epochs, early stopping is applied.
- **`restore_best_weights=True`**: During early stopping, the model's weights corresponding to the best validation loss are restored.

- **`epochs=10`**: The model will be trained for 10 epochs.
- **`batch_size=32`**: 32 images will be used in each training step.
- **`callbacks=[early_stopping]`**: The early stopping callback function is used during training.


````python
from tensorflow.keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)

history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=10,
    batch_size=32,
    callbacks=[early_stopping]
)

````

#Start training multiple models simultaneously

````python
import matplotlib.pyplot as plt
from tensorflow.keras.applications import ResNet50, VGG16, InceptionV3, EfficientNetB0
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping

train_gen = ImageDataGenerator(rescale=1./255, validation_split=0.2)
train_data = train_gen.flow_from_directory(
    train_path,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    subset='training'
)
val_data = train_gen.flow_from_directory(
    train_path,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    subset='validation'
)

def create_model(base_model):
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(128, activation='relu')(x)
    x = Dropout(0.5)(x)
    predictions = Dense(1, activation='sigmoid')(x)
    model = Model(inputs=base_model.input, outputs=predictions)
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

models = {
    'ResNet50': ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3)),
    'VGG16': VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3)),
    'InceptionV3': InceptionV3(weights='imagenet', include_top=False, input_shape=(224, 224, 3)),
    'EfficientNetB0': EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224, 224, 3)),
}

history_dict = {}

early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)

for name, base_model in models.items():
    base_model.trainable = False
    model = create_model(base_model)

    print(f"Starting Training: {name}")
    history = model.fit(
        train_data,
        validation_data=val_data,
        epochs=10,
        callbacks=[early_stopping]
    )

    history_dict[name] = history

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
for name, history in history_dict.items():
    plt.plot(history.history['accuracy'], label=f'{name} - Training Accuracy')
    plt.plot(history.history['val_accuracy'], label=f'{name} - Validation Accuracy')
plt.title('Model traning and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
for name, history in history_dict.items():
    plt.plot(history.history['loss'], label=f'{name} - Traning loss')
    plt.plot(history.history['val_loss'], label=f'{name} - Validation loss')
plt.title('Model traning ve Validation loss')
plt.xlabel('Epoch')
plt.ylabel('loss')
plt.legend()

plt.tight_layout()
plt.show()
````
